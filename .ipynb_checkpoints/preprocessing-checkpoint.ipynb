```python
spell = Speller(fast=True) # init spell check
stop_words = stopwords.words('english') # import stop words
wordLemm = WordNetLemmatizer()  # init lemmatizer

file = open("slang.txt", "r") #import slangs word
slangs = file.readlines()
file.close()
```

```python
def replace_repeated(word):
    if wordnet.synsets(word):
        return word
    repl_word = re.compile(r'(\w*)(\w)\2(\w*)').sub(r'\1\2\3', word)
    if repl_word != word:
        return replace_repeated(repl_word)
    else:
        return repl_word
```

```python
def clean_text(text):
    text = str(text).lower() #lower text
    text = html.unescape(text) #unescape html
    text = re.sub('\[.*?\]', '', text) # remove beackets
    text = re.sub('https?://\S+|www\.\S+', '', text) #remove links
    text = re.sub('\@\S+', '', text) #remove usernames
    text = re.sub('<.*?>+', '', text) #remove <>
    text = ' '.join(replace_repeated(word) for word in text.split(' ')) #reduce length of long words   
    text = ' '.join(contractions.fix(word) for word in text.split(' ')) #fix words
    text = spell(text) # spell check words
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) #remove punctution
    text = re.sub('\n', '', text) # remove new lines
    text = re.sub('\w*\d\w*', '', text) # remove 
    text = re.sub(' +', ' ', text) #remove multi space
    return text.lower().strip() # return and strip
```

```python
def remove_slang(text):
    content = text.split(" ")
    for key, word in enumerate(content):
        word = re.sub('[^a-zA-Z0-9-_.]', '', word)
        for slang in slangs:
            slang = slang.split("=")
            if slang[0].strip().upper() == word.strip().upper():
                content[key] = slang[1].lower().strip()
    return ' '.join(content)
```

```python
def preprocess_data(text):
    # Clean puntuation, urls, and so on
    text = clean_text(text)
    #replace slang words
    if len(text):
        text = remove_slang(text)
    # Remove stopwords
    if len(text):
        text = ' '.join(word for word in text.split(' ') if word not in stop_words)
    # lemmatize all the words in the sentence
    if len(text):
        text = ' '.join(wordLemm.lemmatize(word) for word in text.split(' '))
    return text
```

```python
preprocess_data("I'm so cold")
```

```python
%%time
df['text_clean'] = df['text'].swifter.apply(preprocess_data)
df['text_clean_len'] = [len(t) for t in df.text_clean]
df.head()
```

```python
%%time
wc = WordCloud(
    background_color='white', 
    max_words=2000
)
wc.generate(' '.join(text for text in df.loc[df['sentiment'] == 1, 'text_clean']))
plt.figure(figsize=(18,10))
plt.title('Top words for POSITIVE sentiment', fontdict={'size': 22,  'verticalalignment': 'bottom'})
plt.imshow(wc)
plt.axis("off")
plt.show()
```
